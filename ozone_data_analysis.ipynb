```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
groupedDF = pd.read_excel('./data/ozone-nox-so2-parsed.xlsx',  encoding = "utf-8")
groupedDF["Date"], groupedDF["Time"] = groupedDF["Date"].str.split(' ', 1).str
```

```python
groupedDF = groupedDF.replace('None', "0.0")
groupedDF["Ozone"] = groupedDF["Ozone"].astype(float).fillna(0.0)
groupedDF["NOx"] = groupedDF["NOx"].astype(float).fillna(0.0)
groupedDF["SO2"] = groupedDF["SO2"].astype(float).fillna(0.0)
groupedDF.set_index("Date")
```

```python
groupedDF.plot()
```
