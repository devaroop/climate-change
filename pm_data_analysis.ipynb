```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
groupedDF = pd.read_excel('./data/pm25pm10-parse.xlsx',  encoding = "utf-8")
```

```python
groupedDF["Date"], groupedDF["Time"] = groupedDF["Date"].str.split(' ', 1).str
```

```python
groupedDF = groupedDF.replace('None', "0.0")
groupedDF["PM2.5"] = groupedDF["PM2.5"].astype(float).fillna(0.0)
groupedDF["PM10"] = groupedDF["PM10"].astype(float).fillna(0.0)
groupedDF.set_index("Date")
```

```python
groupedDF.plot()
```
