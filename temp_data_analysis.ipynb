```python
import pandas as pd 
```

```python
df = pd.read_excel('./data/weather_historic.xls',  encoding = "utf-8")
```

```python
df
```

```python
df = df[["Local time in Pune","T"]]
```

```python
df["Date"], df["Daytime"] = df["Local time in Pune"].str.split(' ', 1).str
```

```python
groupDF = df.groupby(['Date']).mean()
```

```python
groupedDF = pd.DataFrame(groupDF)

```

```python
groupedDF.plot.bar(x=['Date'], y=['T'], rot=0)
```
