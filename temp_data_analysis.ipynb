```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('./data/weather_historic.xls',  encoding = "utf-8")
```

```python
df
```

```python
df = df[["Local time in Pune","T"]]
```

```python
df["Date"], df["Daytime"] = df["Local time in Pune"].str.split(' ', 1).str
```

```python
groupDF = df.groupby(['Date']).mean()
```

```python
groupedDF = pd.DataFrame(groupDF).reset_index()

```

```python
groupedDF["Day"], groupedDF["Month"], groupedDF["Year"] = groupedDF["Date"].str.split('.').str
```

```python
groupedDF.reset_index()["Date"].str.split('.').str
```

```python
dayDetailDF = groupedDF[(groupedDF["Month"] == "01") & (groupedDF["Day"] == "01")]
```

```python
dayDetailDF.set_index("Date")
```

```python
dayDetailDF.plot()
```
